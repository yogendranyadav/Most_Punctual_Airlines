# Finding the most punctual airlines

### Objective:
  Here we are to find the most punctual airlines. We will try to use Histograms and Density plots to come to a conlusion.

### Dataset used: 
  Download the data file (formatted_flights.csv) [here](https://github.com/WillKoehrsen/Data-Analysis/tree/master/univariate_dist/data)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing the dataset here:

```python
flights = pd.read_csv('formatted_flights.csv')
flights.head()
```

## Plotting histogram:

To check the arrival delays of all the flights and their distributions.

```python
fig,ax = plt.subplots(figsize = (14,8))
plt.hist(flights['arr_delay'], color = 'teal', edgecolor = 'black', bins = int(180/5) )
# sns.distplot(flights['arr_delay'], hist = True, kde = False, bins = int(180/5), color = 'teal', \
#              hist_kws = {'edgecolor': 'black'})
plt.title('Histogram of arrival delays', fontsize = 18)
plt.xlabel("Delay(mins)", fontsize = 16)
plt.ylabel('Flights', fontsize = 16)
```

### Varying the bin width to see if we can get a better insight out of Histagrams

```python
fig = plt.subplots(figsize = (14,8))
for i, binwidth in enumerate([1,5,10,15]):
    ax = plt.subplot(2, 2, i + 1)
    ax.hist(flights['arr_delay'], color = 'teal', edgecolor = 'black', bins = int(180/binwidth))
    ax.set_title('Histogram  with %d bins' % binwidth, size = 24)
    ax.set_xlabel("Delay(mins)", size = 20)
    ax.set_ylabel('Flights', size = 20)
plt.tight_layout()
plt.show()
```

```python
flights['name'].unique().tolist()
```

## Density plots:

- Moving to Density plots as it was difficult to derive any insight from the histogram.
- Plotting density plot for each airlines arrival delays.

```python
airlines = ['United Air Lines Inc.',
 'American Airlines Inc.',
 'JetBlue Airways',
 'Delta Air Lines Inc.',
 'ExpressJet Airlines Inc.',
 'Envoy Air',
 'US Airways Inc.',
 'Southwest Airlines Co.',
 'Virgin America',
 'AirTran Airways Corporation',
 'Alaska Airlines Inc.',
 'Endeavor Air Inc.',
 'Frontier Airlines Inc.',
 'Hawaiian Airlines Inc.',
 'Mesa Airlines Inc.',
 'SkyWest Airlines Inc.']
fig = plt.subplots(figsize =(10,120))
for i, airline in enumerate(airlines):
    ax = plt.subplot(len(airlines),1, i+1)
    subset = flights[flights['name'] == airline]
    sns.distplot(subset['arr_delay'], hist = False, kde = True,rug = True, label = airline, kde_kws = {'linewidth': 3})
    ax.set_title('Density plot for %s' % airline, size = 20)
    ax.set_xlabel("Delay(mins)", size = 18)
    ax.set_ylabel('Density', size = 18)
    # fix the dimensions for easy comparability
    ax.set(ylim=(0, 0.0275)) 
plt.show()
```

**By seeing the density plots with rug, we can say that ALASKA airlines and Hawaiian airlines are more punctual than the others**

## Shaded Density plot for Alaska and Hawaiian Airlines

```python
airlines = ['Hawaiian Airlines Inc.',
 'Alaska Airlines Inc.']
fig = plt.subplots(figsize =(16,10))
for airline in airlines:
    subset = flights[flights['name'] == airline]
    sns.distplot(subset['arr_delay'], hist = False, kde = True,rug = True, label = airline, kde_kws = {'linewidth': 3, 'shade': True})
plt.legend(title = 'Airlines', fontsize = 12)
plt.title('Density Plot with Multiple airlines', fontsize = 18)
plt.xlabel('Delay(mins)', fontsize = 16)
plt.ylabel('Density', fontsize = 16)
plt.show()
```

## Conclusion:

- The Top two punctual airlines are Hawaiian Airlines Inc and Alaska Airline Inc.
- The next time, you want to be punctual then you know which airlines to go for.
